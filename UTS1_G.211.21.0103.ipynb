```python
from sklearn.model_selection import train_test_split  # Mengimpor fungsi train_test_split untuk membagi data menjadi set train & testing
from sklearn.neighbors import KNeighborsClassifier  # Mengimpor kelas KNeighborsClassifier untuk algoritma K-Nearest Neighbors (K-NN)
from sklearn.metrics import accuracy_score  # Mengimpor fungsi accuracy_score untuk mengukur akurasi model
from sklearn.preprocessing import StandardScaler  # Mengimpor kelas StandardScaler untuk penskalaan fitur
```

```python
categorical_columns = ['Sex', 'Obesity', 'DLP', 'WPE', 'Atypical', 'Nonanginal', 'LVH', 'PRP'] # Mendefinisikan kategori kolom
CAD = pd.get_dummies(CAD, columns=categorical_columns, drop_first=True) # Mengubah kolom-kolom kategorikal menjadi variabel dummy

# Memisahkan fitur (X) dan target (y)
X = CAD.drop(columns=['CAD'])  # X berisi semua kolom kecuali 'CAD' sebagai target
y = CAD['CAD']  # y berisi kolom 'CAD' sebagai target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50) # Membuat set train & testing 20/80
```

```python
# Normalisasi data
scaler = StandardScaler()  # Membuat objek StandardScaler untuk normalisasi data
X_train = scaler.fit_transform(X_train)  # Melakukan normalisasi data pada set train
X_test = scaler.transform(X_test)  # Melakukan normalisasi data pada set pengujian

# Buat dan latih model KNN dengan n_neighbors = 5
knn = KNeighborsClassifier(n_neighbors=5)  # Membuat objek KNeighborsClassifier dengan 5 tetangga terdekat
knn.fit(X_train, y_train)  # Melatih model KNN menggunakan data train

# Evaluasi model KNN
y_pred = knn.predict(X_test)  # Memprediksi label/target pada data pengujian
accuracy = accuracy_score(y_test, y_pred)  # Mengukur akurasi model dengan membandingkan hasil prediksi dengan label sebenarnya
print("Akurasi model KNN:", accuracy)  # Mencetak akurasi model KNN
```
