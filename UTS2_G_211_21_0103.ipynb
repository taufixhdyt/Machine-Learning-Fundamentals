```python
# Menggantikan isi pada kolom Company agar tidak terjadi duplikat
company_name_mapping = {
    'maxda': 'mazda',
    'Nissan': 'nissan',
    'porcshce': 'porsche',
    'toyouta': 'toyota',
    'vokswagen': 'volkswagen',
    'vw': 'volkswagen'
}

Car['Company'].replace(company_name_mapping, inplace=True)
```

#Visualisasi Data

```python
# Mengimpor library Plotly Express dengan alias px
import plotly.express as px

company_prices = Car.groupby("Company")["price"].mean().reset_index() # Menghitung rata-rata harga mobil untuk setiap perusahaan menggunakan groupby
fig = px.bar(company_prices, x="Company", y="price", title="Average Car Prices by Company") # Membuat diagram batang dengan Plotly Express
fig.update_xaxes(categoryorder="total descending") # Mengatur urutan kategori pada sumbu x berdasarkan total harga secara menurun
fig.update_layout(xaxis_title="Company", yaxis_title="Average Price") # Mengupdate layout diagram dengan judul sumbu x dan y

# Menampilkan diagram
fig.show()
```

```python
# Import library matplotlib dan seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat objek figure dengan ukuran 15x20
fig = plt.figure(figsize=(15, 20))

# Subplot 1: Count plot untuk distribusi fueltype
plt.subplot(3, 2, 1)
sns.countplot(data=Car, x='fueltype')  # Membuat count plot distribusi fueltype mobil dari data 'Car'
plt.title('Car fueltype distribution plot') # Menambahkan judul pada subplot

# Subplot 2: Count plot untuk distribusi fuelsystem
plt.subplot(3, 2, 2)
sns.countplot(data=Car, x='fuelsystem') # Membuat count plot distribusi fuelsystem mobil dari data 'Car'
plt.title('Car fuelsystem plot') # Menambahkan judul pada subplot

# Subplot 3: Count plot untuk distribusi aspiration
plt.subplot(3, 2, 3)
sns.countplot(data=Car, x='aspiration') # Membuat count plot aspiration mobil dari data 'Car'
plt.title('Car aspiration plot') # Menambahkan judul pada subplot

# Subplot 4: Count plot untuk distribusi doornumber
plt.subplot(3, 2, 4)
sns.countplot(data=Car, x='doornumber') # Membuat count plot doornumber mobil dari data 'Car'
plt.title('Car doornumber plot') # Menambahkan judul pada subplot

# Subplot 5: Count plot untuk distribusi carbody
plt.subplot(3, 2, 5)
sns.countplot(data=Car, x='carbody') # Membuat count plot carbody mobil dari data 'Car'
plt.title('Car carbody plot') # Menambahkan judul pada subplot

# Subplot 6: Count plot untuk distribusi drivewheel
plt.subplot(3, 2, 6)
sns.countplot(data=Car, x='drivewheel') # Membuat count plot drivewheel mobil dari data 'Car'
plt.title('Car drivewheel plot') # Menambahkan judul pada subplot

# Menampilkan plot
plt.show()
```

```python
correlation_matrix = Car.corr(numeric_only=True) # Menghitung matriks korelasi antar fitur numerik dalam DataFrame 'Car'
plt.figure(figsize=(12, 10)) # Mengatur ukuran gambar untuk heatmap (12x10)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5) # Membuat heatmap dengan seaborn untuk visualisasi matriks korelasi
plt.title("Correlation Heatmap") # Menambahkan judul ke heatmap
plt.show() # Menampilkan heatmap
```

Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Pilih fitur-fitur yang akan digunakan untuk membangun model
features = ['symboling', 'wheelbase', 'carlength', 'carwidth', 'curbweight']
X = Car[features]

# Pilih kolom target
y = Car['fueltype']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30) # Memisahkan data menjadi data pelatihan dan pengujian
model = DecisionTreeClassifier(max_depth=5, min_samples_leaf=10) # Membuat model Decision Tree
model.fit(X_train, y_train) # Melatih model
y_pred = model.predict(X_test) # Melakukan prediksi pada Car pengujian

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi:", accuracy) # Menampilkan data
```
